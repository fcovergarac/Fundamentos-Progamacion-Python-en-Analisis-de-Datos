### Definición de clase **Contacto**

- Se incluyen docstring y comentarios para facilitar el entendimiento el código
para otros posibles colaboradores del proyecto.

- Incluye prueba del funcionamiento instanciando la clase (creando un objeto), antes de utilizarla dentro de la función principal final.

```python
# Primero se crea la clase contacto.
class Contacto:
  '''
  ¿Qué hace?
  Clase que representa un contacto dentro de la agenda con sus datos.

  Atributos:
  nombre (str): Nombre del contacto.
  telefono (str): Número de teléfono del contacto.
  email (str): Dirección de correo electrónico del contacto.
  direccion (str): Dirección física del contacto.
  '''
  def __init__(self, nombre, telefono, email, direccion):
    # Inicialización de los atributos del contacto con sus datos proporcionados.
    # dado que el input por defecto utilizar str, quedaran todos los datos cómo tal (incluyendo teléfono, ya que para no se harán operaciones matemáticas con él)
    self.nombre = nombre
    self.telefono = telefono
    self.email = email
    self.direccion = direccion

  def convertir_diccionario(self):
    # Método convierte los atributos del contacto en un diccionario.
    '''
    ¿Qué hace?
    Convierte los atributos del contacto en un diccionario.

    Retorna:
    Un diccionario con los atributos del contacto.
    '''
    return {
        "nombre": self.nombre,
        "telefono": self.telefono,
        "email": self.email,
        "direccion": self.direccion
    }

  def __str__(self):
    # Método devuelve string con los datos del contacto.
    '''
    ¿Qué hace?
    Devuelve un string con los datos del contacto.
    '''
    return (f"Nombre: {self.nombre} | "
            f"Teléfono: {self.telefono} | "
            f"Email: {self.email} | "
            f"Dirección: {self.direccion}")

# Este codigo no lo utilizaré porque me gustó más la opción lineal del __str__ separado por pipe
#    return f"Nombre: {self.nombre}\nTeléfono: {self.telefono}\nEmail: {self.email}\nDirección: {self.direccion}"
```

```python
# PRUEBA ->>>>>>>> de uso de docstrings para visualizar lo que recibe la clase Contacto, lo que hace método convertir_diccionario y su método str..

print("Docstring de clase Contacto:")
print(Contacto.__doc__)

print()
print("Docstring de metodo convertir_diccionario de clase Contacto:")
print(Contacto.convertir_diccionario.__doc__)

print()
print("Docstring de metodo __str__ de clase Contacto:")
print(Contacto.__str__.__doc__)
```

```python
# PRUEBA ->>>>>>>> de la clase Contacto, instanciando la clase (creando objetos (3 ejemplos))..

print("Creación de 3 contactos usando la clase Contacto: \n")
print()

contacto_1 = Contacto("Javiera", "+56 9 12345678", "javiera@bootcamp.com", "Calle Cibernetica 123")
print(f"--> Los datos del contacto son: \n")
print(contacto_1)

print()

contacto_2 = Contacto("Juan", "+56 9 87654321", "juan@bootcamp.com", "Calle Data Science 456")
print(f"--> Los datos del contacto son: \n")
print(contacto_2)

print()

contacto_3 = Contacto("Ana", "+56 9 55555555", "ana@bootcamp.com", "Calle Machine Learning 7")
print(f"--> Los datos del contacto son: \n")
print(contacto_3)
```

### Definición de clase **Agenda**

- Se incluyen docstring y comentarios para facilitar el entendimiento el código
para otros posibles colaboradores del proyecto.

- Incluye prueba del funcionamiento instanciando la clase (creando un objeto), antes de utilizarla dentro de la función principal final.

```python
# se crea la clase agenda, con su inicializador y funciones que cumplirá (agregar, listar, buscar, eliminar y editar contactos)

class Agenda:
  '''
  ¿Qué hace?
  Clase que representa una agenda de contactos con una lista de contactos.

  Atributos:
  contactos (list): Lista de contactos en la agenda.
  '''

  def __init__(self):
    # Inicialización de la agenda con una lista vacía de contactos.
    self.contactos = [] # Es lista guardará diccionarios dentro a partir de método de la clase contacto que los transforma como tal (en dict)

  def agregar_contacto(self, contacto):
    '''
    ¿Qué hace?
    Agrega un contacto a la agenda.

    Parámetros:
    contacto (Contacto): Contacto a agregar a la agenda.
    '''
    # Agrega un contacto a la agenda.
    self.contactos.append(contacto.convertir_diccionario())

  def listar_contactos(self):
    '''
    ¿Qué hace?
    Lista todos los contactos en la agenda.
    '''
    for c in self.contactos:
      # cada 'c' es un diccionario, así que debemos iterar dentro de la lista llamandolo con el nombre de la clave..
      print(f"Nombre: {c['nombre']} | Teléfono: {c['telefono']} | Email: {c['email']} | Dirección: {c['direccion']}")

  def buscar_contacto(self, nombre):
    '''
    ¿Qué hace?
    Busca un contacto por su nombre en la agenda.

    Parámetros:
    nombre (str): Nombre del contacto a buscar.

    Retorna:
    Contacto: Si se encuentra el contacto en la agenda, devuelve el contacto.

    None: Si no se encuentra el contacto en la agenda.
    '''
    for c in self.contactos:
      if c["nombre"].lower() == nombre.lower():
        return f"Contacto '{nombre}' encontrado! Sus datos con los siguientes:\n Nombre: {c['nombre']} | Teléfono: {c['telefono']} | Email: {c['email']} | Dirección: {c['direccion']}"
    return None

  def eliminar_contacto(self, nombre):
    '''
    ¿Qué hace?
    Elimina un contacto por su nombre de la agenda.

    Parámetros:
    nombre (str): Nombre del contacto a eliminar.

    Retorna:
    bool: True si el contacto se eliminó correctamente, False si no se encontró el contacto.
    '''
    for c in self.contactos:
      if c["nombre"].lower() == nombre.lower():
        self.contactos.remove(c)
        return True
    return False

  def editar_contacto(self, nombre, telefono=None, email=None, direccion=None):
    '''
    ¿Qué hace?
    Edita un contacto por su nombre en la agenda.

    Parámetros:
    nombre (str): Nombre del contacto a editar
    telefono (str, opcional): Nuevo número de teléfono del contacto.
    email (str, opcional): Nueva dirección de correo electrónico del contacto.
    direccion (str, opcional): Nueva dirección física del contacto.

    Retorna:
    bool: True si el contacto se editó correctamente, False si no se encontró el contacto.
    '''
    for c in self.contactos:
      if c["nombre"].lower() == nombre.lower():
        #ACtualiza solo los campos que se entreguen (sserán opcionales, por eso el None en los parámetros del método)
        if telefono:
          c["telefono"] = telefono
        if email:
          c["email"] = email
        if direccion:
          c["direccion"] = direccion
        return True

    return False
```

```python
# PRUEBA ->>>>>>>> de la clase agenda. Se agregan los 3 contactos creados como prueba anteriormente contacto_1 contacto_2 y contacto_3

agenda = Agenda()
agenda.agregar_contacto(contacto_1)
agenda.agregar_contacto(contacto_2)
agenda.agregar_contacto(contacto_3)

# Listar o mostrar los contactos agregados en la agenda..
print("--> Lista de contactos en la agenda:")
agenda.listar_contactos()

# Buscar un contacto
print()

print("Ejemplo, busqueda de JAviEra:")
agenda.buscar_contacto("JAviEra") # A proposito puse mayusculas entremedio, ya que el método contiene un if que los 'loweriza'

# Eliminar un contacto
print()
print(f"Eliminando a Javiera de la agenda...")
if agenda.eliminar_contacto("Javiera"):
  print("Contacto eliminado correctamente.")
else:
  print("Contacto no encontrado.")

#Listar again para revisar la agenda actualizada
print()
print("--> Lista de contactos en la agenda:")
agenda.listar_contactos()

# Actualizar un contacto
print()
print(f"Actualizando a Ana de la agenda...")
if agenda.editar_contacto("Ana", telefono="+56 9 95456456", direccion="Calle Data Science 789"): # email deberá seguir siendo el mismo ya que no se le pasa para actualizar
  print("Contacto actualizado correctamente.")
else:
  print("Contacto no encontrado.")

#Listar again para revisar la agenda actualizada
print()
print("--> Lista de contactos en la agenda:")
agenda.listar_contactos()
```

### Función Menú interactivo

El usuario final que desea gestionar sus contactos de clientes no necesita visualizar nuestro código o utilizar las clases directamente, por lo que se crea el siguiente menú interactivo para facilitar la manera en que interactura con el sistema.

```python
# Todo el menú será parte de una función de python que luego en una celda individual se podrá ejecutar y se verá de manera limpia su ejecución de cara al usuario final

def menu():
  '''
  función que muestra menú interactivo para gestionar la agenda y contactos de clientes.
  Permite al usuario elegir entre opciones de agregar, listar, buscar, eliminar, editar o salir.
  '''
  # se instancia clase agenda para crear objeto mi_agenda vacía
  mi_agenda = Agenda()

  # bucle while que preguntará al usuario de manera indefinida hasta que decida salir (ingresando 6)
  while True:
    # opciones disponibles
    print("\n --- Menú de Agenda ---")
    print("1. Agregar contacto")
    print("2. Listar contactos")
    print("3. Buscar contacto")
    print("4. Eliminar contacto")
    print("5. Editar contacto")
    print("6. Salir")

    # input para obtener la opción elegida por el usuaurio

    opcion = input("Seleccione una opción(1-6): ")

    #opción 1 --> agregar un contacto a la agenda
    if opcion == "1":
      nombre = input("Ingrese el nombre del contacto: ")
      telefono = input("Ingrese el número de teléfono del contacto: ")
      email = input("Ingrese la dirección de correo electrónico: ")
      direccion = input("Ingrese la dirección física del contacto: ")
      # se crea objeto nuevo contacto con sus respectivos parámetros, luego se le pasa al método agregar_contacto de mi_agenda.
      nuevo_contacto = Contacto(nombre, telefono, email, direccion)
      mi_agenda.agregar_contacto(nuevo_contacto)
      print("Contacto agregado correctamente.")

    #opción 2 --> listar los contactos de la agenda
    elif opcion == "2":
      print("\nLista de contactos en la agenda:")
      mi_agenda.listar_contactos()

    #opción 3 --> buscar contacto en la agenda
    elif opcion == "3":
      nombre = input("Ingrese el nombre del contacto:")
      resultado = mi_agenda.buscar_contacto(nombre)
      if resultado:
        print("Contacto encontrado: ", resultado)
      else:
        print("Contacto no encontrado.")

    # opción 4 --> elimnar contacto de la agenda
    elif opcion == "4":
      nombre = input("Ingrese el nombre del contacto a eliminar: ")
      if mi_agenda.eliminar_contacto(nombre):
        print("Contacto eliminado correctamente.")
      else:
        print("Contacto no encontrado.")

    # opción 5 --> actualizar contacto ya guardado en la agenda
    elif opcion == "5":
      nombre = input("Ingrese el nombre del contacto a editar: ")
      telefono = input("Ingrese el nuevo número de teléfono (dejar vacío si no desea cambiarlo): ")
      email = input("Ingrese la nueva dirección de correo electrónico (dejar vacío si no desea cambiarla): ")
      direccion = input("Ingrese la nueva dirección física (dejar vacío si no desea cambiarla): ")

      # si el usuario deja vacío alguno de los campos se pasan como un parametro none
      telefono = telefono if telefono else None
      email = email if email else None
      direccion = direccion if direccion else None

      if mi_agenda.editar_contacto(nombre, telefono, email, direccion):
        print("Contacto actualizado correctamente.")
      else:
        print("Contacto no encontrado.")

    #opción 6 --> salir del programa
    elif opcion == "6":
      print("Saliendo de su sistema de agenda... Que tengas buen día!")
      break

    # opción inválida en el menú. No se inrgesó unos de las opción disponibles (1-6)
    else:
      print("Opción inválida. Por favor, seleccione una opción válida (1-6).")

```

  - Editar Contactos



```python
# Ejecución de función menú que utilizar las clases Contacto y Agenda para operar, donde el usuario podrá interactuar con su agenda y contactos.

#

print("> Bienvenid@ al sistema de Gestión de contactos <\n")
print("A continuación se desplagará menú interactivo para gestiones tu agenda (y contactos)...")

menu()

print("> Sesión Finalizada <")
```
